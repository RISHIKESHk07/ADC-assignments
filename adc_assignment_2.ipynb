# Simulation of PDF from Two Gaussian Random Variables

## Objectives:

We will be using X & Y gaussian rv

1. Verify that **R follows a Rayleigh distribution**
2. Verify that **Z follows an Exponential distribution**
3. Plot histograms and compare with theoretical PDFs

---

##Rayleigh Distribution for $$( R = \sqrt{X^2 + Y^2} )$$

Theoretical PDF:

$$
f_R(r) = \frac{r}{\sigma^2} e^{-r^2/(2\sigma^2)}
$$

## Exponential Distribution for \( Z = R^2 \)

Theoretical PDF:

$$
f_Z(z) = \frac{1}{\omega} e^{-z/\omega}
$$

where:

$$
\omega = E[R^2] = 2\sigma^2
$$

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh, expon

# Parameters (from cell 5d8c7fb9)
sigma = 2
N = 100000

# Generate Gaussian random variables (from cell 5d8c7fb9)
X = np.random.normal(0, sigma, N)
Y = np.random.normal(0, sigma, N)

# Compute R (from cell e9f7cc21)
R = np.sqrt(X**2 + Y**2)

# Theoretical Rayleigh distribution (from cell e9f7cc21)
r_vals = np.linspace(0, np.max(R), 500)
pdf_rayleigh = (r_vals / sigma**2) * np.exp(-r_vals**2 / (2*sigma**2))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Histogram of R
axes[0].hist(R, bins=100, density=True, alpha=0.7, color='skyblue', label='Simulated R')
axes[0].set_title('Histogram of R (Simulated)')
axes[0].set_xlabel('R')
axes[0].set_ylabel('Probability Density')
axes[0].legend()

# Plot 2: Theoretical Rayleigh PDF
axes[1].plot(r_vals, pdf_rayleigh, color='red', linestyle='--', label='Theoretical PDF')
axes[1].set_title('Theoretical Rayleigh PDF')
axes[1].set_xlabel('R')
axes[1].set_ylabel('Probability Density')
axes[1].legend()

plt.suptitle('Comparison: Simulated R vs. Theoretical Rayleigh Distribution')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh, expon

# Parameters (from cell 5d8c7fb9)
sigma = 2
N = 10000000

# Generate Gaussian random variables (from cell 5d8c7fb9)
X = np.random.normal(0, sigma, N)
Y = np.random.normal(0, sigma, N)

# Compute R (from cell e9f7cc21)
R = np.sqrt(X**2 + Y**2)

# Compute Z (from cell ae4d4bbd)
Z = R**2

# Omega = E[R^2] (Using theoretical value as per markdown cell 4a83681f)
Omega = 2 * sigma**2

z_vals = np.linspace(0, np.max(Z), 500)
pdf_exp = (1/Omega) * np.exp(-z_vals / Omega)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Histogram of Z
axes[0].hist(Z, bins=100, density=True, alpha=0.7, color='lightgreen', label='Simulated Z')
axes[0].set_title('Histogram of Z (Simulated)')
axes[0].set_xlabel('Z = R')
axes[0].set_ylabel('Probability Density')
axes[0].legend()

# Plot 2: Theoretical Exponential PDF
axes[1].plot(z_vals, pdf_exp, color='blue', linestyle='--', label='Theoretical PDF')
axes[1].set_title('Theoretical Exponential PDF')
axes[1].set_xlabel('Z = R')
axes[1].set_ylabel('Probability Density')
axes[1].legend()

plt.suptitle('Comparison: Simulated Z vs. Theoretical Exponential Distribution')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
